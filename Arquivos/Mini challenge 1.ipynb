```python
import pandas as pd
import random as rd 

# O objetivo é dividir 19 pessoas em 4 grupos, sendo estes compostos por 3 grupos de 5 pessoas e 1 grupo de 4 pessoas.
## No entanto, 4 pessoas possuem restrição de não estarem no mesmo grupo, logo essas devem ser divididas 1 para cada grupo.

# Leitura do arquivo .xlsx contendo os nomes e suas eventuais restrições
df = pd.read_excel("C:/Users/laupo/OneDrive/Área de Trabalho/vscode/Mini challenge - Semana 1.xlsx")

# Criação das listas base para divisão das pessoas. Aqui inseri manualmente a quantidade de grupos a se dividir.
grupo_restrito = []
grupo_livre = []
separacao_grupos = {"Grupo 1":"","Grupo 2":"","Grupo 3":"","Grupo 4":""}

# Separar em grupos controle de acordo com as restrições (0 = livre ; 1 = restrito) 
for i,nome in enumerate(df['Nome']):
    if df['Restrição'][i] == 1:
        grupo_restrito.append(nome)
    else:
        grupo_livre.append(nome)

# Divisão dos grupos
## Iteração responsável por dividir os grupos
for i,grupo in enumerate(separacao_grupos):
    
    # Definição de uma lista auxiliar para controle de repetições
    sorteados = []
    
    # Sorteio de um item aleatório dentro do grupo restrição (após a escolha, o item é removido da lista original)
    item_restrito = rd.choice(grupo_restrito)
    sorteados.append(item_restrito)
    grupo_restrito.remove(item_restrito)

    # Sorteio de um item aleatório dentro do grupo livre, respeitando o tamanho da amostra (mesma observação do grupo restrito)
    ## Divisão do grupo de 5 pessoas
    if i < 3:
        item_livre = rd.sample(grupo_livre,k=4)
        sorteados += item_livre
        for item in item_livre:
            grupo_livre.remove(item)
            
    ## Divisão do grupo de 4 pessoas   
    else:
        item_livre = rd.sample(grupo_livre,k=3)
        sorteados += item_livre
        for item in item_livre:
            grupo_livre.remove(item)
    
    # Atribuição dos grupos
    separacao_grupos[grupo] = sorteados

# A partir daqui farei apenas uma modificação na exibição dos dados para melhor visualização do output
mensagem = ""
# Criei um looping para exibir uma mensagem no formato : "Grupo 1 é compostos pelos membros : lista ordenada dos membros"
for x,lista in enumerate(separacao_grupos.values()):
    
    for i,item in enumerate(lista):
        
        # A mensagem agrega dois fatores, a primeira parte que identifica o grupo e as subsequentes com seus membros.
        if i == 0:
            mensagem += f'''
            • O grupo {x+1} é composto pelos seguintes membros: {item}
            '''
        elif i > 0:
            mensagem += f'''
                                                            {item}
            '''
    # Aqui uma exibição de um separador para facilitar a visualização dos grupos
    mensagem += "-"*100
    
print(mensagem)
```
